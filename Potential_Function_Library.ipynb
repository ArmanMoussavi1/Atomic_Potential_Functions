##**1 LENNARD-JONES POTENTIAL**

Lennard, J. E., and I. Jones. "On the determination of molecular fields.â€”I. From the variation of the viscosity of a gas with temperature." Proceedings of the Royal Society of London. Series A, containing papers of a mathematical and physical character 106.738 (1924): 441-462.


\begin{equation}
U_{ij}^{LJ} = 4 \epsilon \left(\left(\frac{\sigma}{r}\right)^{12} - \left(\frac{\sigma}{r}\right)^6\right)
\end{equation}
<br>

**$U_{ij}^{LJ}$** = intermolecular potential between the two atoms or molecules

**$\epsilon$** = well depth and a measure of how strongly the two particles attract each other

**$\sigma$** = distance at which the intermolecular potential between the two particles is zero.  $\sigma$ gives a measurement of how close two nonbonding particles can get and is thus referred to as the van der Waals radius. It is equal to one-half of the internuclear distance between nonbonding particles

**$r$** = distance of separation between both particles (measured from the center of one particle to the center of the other particle)

```python
#@title 1.1 LJ Potential


#---------- Input ----------#

poten_func = 'Lennard-Jones'

epsilon = 0.36
sigma = 2.88


r = np.linspace(1*10**-10, 20, 10000)

#---------------------------#


def U_LJ(epsilon, sigma, r):
    return 4 * epsilon * ((sigma / r)**12 - (sigma / r)**6)



print('\n')
print('------------------------------')
print('\n')
print('Well Depth:', epsilon)
print('\n')
print('Equilibrium Distance:', sigma)
print('\n')
print('------------------------------')
print('\n')



potential = U_LJ(epsilon, sigma, r)


plt.plot(r, potential, 'b')
plt.xlabel('r')
plt.ylabel('U[\u03B5, \u03C3, r]')
plt.title(f'{poten_func} Potential')
plt.xlim(0, 20)
plt.ylim(-0.5, 1)
plt.grid(True)
plt.show()
```

```python
#@title 1.2 Comparing LJ Potential Paramters


#---------- Input ----------#

poten_func = 'Lennard-Jones'

epsilon1 = 0.36  # Well Depth 1
sigma1 = 2.88  # Equilibrium Distance 1


epsilon2 = 0.4  # Well Depth 2
sigma2 = 3.0  # Equilibrium Distance 2


r = np.linspace(1*10**-10, 20, 10000)

#---------------------------#


def U_LJ(epsilon, sigma, r):
    return 4 * epsilon * ((sigma / r)**12 - (sigma / r)**6)





print('\n')
print('------------------------------')
print('\n')
print('U:', 1)
print('\n')
print('Well Depth:', epsilon1)
print('\n')
print('Equilibrium Distance:', sigma1)
print('\n')
print('\n')
print('U:', 2)
print('\n')
print('Well Depth:', epsilon2)
print('\n')
print('Equilibrium Distance:', sigma2)
print('\n')
print('------------------------------')
print('\n')



potential1 = U_LJ(epsilon1, sigma1, r)
potential2 = U_LJ(epsilon2, sigma2, r)


plt.plot(r, potential1, 'b', label='U 1')
plt.plot(r, potential2, 'r', label='U 2')
plt.xlabel('r')
plt.ylabel('U[\u03B5, \u03C3, r]')
plt.title(f'Comparing {poten_func} Potential Paramters')
plt.xlim(0, 20)
plt.ylim(-0.5, 1)
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

##**2 DOUBLE EXPONENTIAL POTENTIAL**

Wu, Xiongwu, and Bernard R. Brooks. "A double exponential potential for van der Waals interaction." AIP advances 9.6 (2019).

\begin{equation}
U_{ij}^{DE} = \epsilon \left(\frac{\beta e^{\alpha}}{\alpha - \beta} \exp\left(-\alpha \frac{r}{\sigma}\right) - \frac{\alpha e^{\beta}}{\alpha - \beta} \exp\left(-\beta \frac{r}{\sigma}\right)\right)
\end{equation}
<br>



**$U_{ij}^{DE}$** = intermolecular potential between the two atoms or molecules

$\alpha$ = steepness of the repulsive interation

$\beta$ = decay of the attraction

>>$U_{ij}^{DE}$ = $U_{ij}^{LJ}$ when $\alpha$ = 17.47 ; $\beta$ = 4.1

**$\epsilon$** = well depth and a measure of how strongly the two particles attract each other

**$\sigma$** = distance at which the intermolecular potential between the two particles is zero.  $\sigma$ gives a measurement of how close two nonbonding particles can get and is thus referred to as the van der Waals radius. It is equal to one-half of the internuclear distance between nonbonding particles

**$r$** = distance of separation between both particles (measured from the center of one particle to the center of the other particle)

```python
#@title 2.1 DE Potential

#---------- Input ----------#

poten_func = 'Double Exponential'

alpha = 17.47
beta = 4.1
sigma = 2.88
epsilon = 0.36

r = np.linspace(1*10**-10, 20, 10000)

#---------------------------#


def U_DE(alpha, beta, sigma, epsilon, r):
    return epsilon * ((beta * np.exp(alpha)) / (alpha - beta) * np.exp(-alpha * r / sigma) -
                      (alpha * np.exp(beta)) / (alpha - beta) * np.exp(-beta * r / sigma))




print('\n')
print('------------------------------')
print('\n')
print('Steepness of the repulsive interation:', alpha)
print('\n')
print('Decay of the attraction:', beta)
print('\n')
print('Well Depth:', epsilon)
print('\n')
print('Equilibrium Distance:', sigma)
print('\n')
print('------------------------------')
print('\n')


potential =U_DE(alpha, beta, sigma, epsilon, r)


plt.plot(r, potential, 'b')
plt.xlabel('r')
plt.ylabel('U[\u03B1, \u03B2, \u03B5, \u03C3, r]')
plt.title(f'{poten_func} Potential')
plt.xlim(0, 20)
plt.ylim(-0.5, 1)
plt.grid(True)
plt.show()
```

```python
#@title 2.1 Comparing DE Potential Parameters

#---------- Input ----------#

poten_func = 'Double Exponential'

alpha1 = 17.47
beta1 = 4.1
sigma1 = 2.88
epsilon1 = 0.36


alpha2 = 18
beta2 = 5
sigma2 = 3
epsilon2 = 0.4


r = np.linspace(1*10**-10, 20, 10000)

#---------------------------#



def U_DE(alpha, beta, sigma, epsilon, r):
    return epsilon * ((beta * np.exp(alpha)) / (alpha - beta) * np.exp(-alpha * r / sigma) -
                      (alpha * np.exp(beta)) / (alpha - beta) * np.exp(-beta * r / sigma))




print('\n')
print('------------------------------')
print('\n')
print('U:', 1)
print('\n')
print('Steepness of the repulsive interation:', alpha1)
print('\n')
print('Decay of the attraction:', beta1)
print('\n')
print('Well Depth:', epsilon1)
print('\n')
print('Equilibrium Distance:', sigma1)
print('\n')
print('\n')
print('U:', 2)
print('\n')
print('Steepness of the repulsive interation:', alpha2)
print('\n')
print('Decay of the attraction:', beta2)
print('\n')
print('Well Depth:', epsilon2)
print('\n')
print('Equilibrium Distance:', sigma2)
print('\n')
print('------------------------------')
print('\n')




potential1 =U_DE(alpha1, beta1, sigma1, epsilon1, r)
potential2 =U_DE(alpha2, beta2, sigma2, epsilon2, r)


plt.plot(r, potential1, 'b', label='U 1')
plt.plot(r, potential2, 'r', label='U 2')
plt.xlabel('r')
plt.ylabel('U[\u03B1, \u03B2, \u03B5, \u03C3, r]')
plt.title(f'Comparing {poten_func} Potential Paramters')
plt.xlim(0, 20)
plt.ylim(-0.5, 1)
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

##**3 MORSE POTENTIAL**

Morse, Philip M. "Diatomic molecules according to the wave mechanics. II. Vibrational levels." Physical review 34.1 (1929): 57.

<br>
NOTES:

Provides a better description for the potential energy of a diatomic molecule than the quantum harmonic oscillator

\begin{equation}
U_{ij}^{M} = D_e \left(1 - e^{-\alpha \left(r - r_e \right)} \right)^{2}
\end{equation}
<br>



**$U_{ij}^{M}$** = intermolecular potential between the two atoms or molecules

$D_e$ = well depth related to the molecule dissociation energy

$\alpha$ = inverse length parameter related to the curvature of the potential at the origin (controls the width of the potential)

$r_e$ = equilibrium bond distance

$r$ = distance between atoms


```python
#@title 3.1 Morse Potential


#---------- Input ----------#

poten_func = 'Morse'

D_e = 0.36
alpha = 3
r_e = 2.88


r = np.linspace(1*10**-10, 20, 10000)

#---------------------------#


def U_M(alpha, D_e, r, r_e):
    return D_e * (1 - np.exp(-alpha * (r - r_e)))**2



print('\n')
print('------------------------------')
print('\n')
print('Well depth:', D_e)
print('\n')
print('inverse length parameter:', alpha)
print('\n')
print('Equilibrium bond distance:', r_e)
print('\n')
print('------------------------------')
print('\n')



potential = U_M(alpha, D_e, r, r_e)


plt.plot(r, potential, 'b')
plt.xlabel('r')
plt.ylabel('U[\u03B1, $D_e$, $r_e$, $r$]')
plt.title(f'{poten_func} Potential')
plt.xlim(0, 20)
plt.ylim(-0.5, 1)
plt.grid(True)
plt.show()
```

```python
#@title 3.2 Comparing Morse Potential Paramters


#---------- Input ----------#

poten_func = 'Morse'

D_e1 = 0.36
alpha1 = 3
r_e1 = 2.88


D_e2 = 0.4
alpha2 = 3.5
r_e2 = 3


r = np.linspace(1*10**-10, 20, 10000)

#---------------------------#


def U_M(alpha, D_e, r, r_e):
    return D_e * (1 - np.exp(-alpha * (r - r_e)))**2





print('\n')
print('------------------------------')
print('\n')
print('U:', 1)
print('\n')
print('Well depth:', D_e1)
print('\n')
print('inverse length parameter:', alpha1)
print('\n')
print('Equilibrium bond distance:', r_e1)
print('\n')
print('\n')
print('U:', 2)
print('\n')
print('Well depth:', D_e2)
print('\n')
print('inverse length parameter:', alpha2)
print('\n')
print('Equilibrium bond distance:', r_e2)
print('\n')
print('------------------------------')
print('\n')



potential1 = U_M(alpha1, D_e1, r, r_e1)
potential2 = U_M(alpha2, D_e2, r, r_e2)


plt.plot(r, potential1, 'b', label='U 1')
plt.plot(r, potential2, 'r', label='U 2')
plt.xlabel('r')
plt.ylabel('U[\u03B1, $D_e$, $r_e$, $r$]')
plt.title(f'Comparing {poten_func} Potential Paramters')
plt.xlim(0, 20)
plt.ylim(-0.5, 1)
plt.legend(loc='best')
plt.grid(True)
plt.show()
```
