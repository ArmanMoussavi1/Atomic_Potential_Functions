##1.1 LENNARD JONES POTENTIAL

\begin{equation}
U_{ij}^{LJ} = 4 \epsilon \left(\left(\frac{\sigma}{r}\right)^{12} - \left(\frac{\sigma}{r}\right)^6\right)
\end{equation}
<br>

**$U_{ij}^{LJ}$** = intermolecular potential between the two atoms or molecules

**$\epsilon$** = well depth and a measure of how strongly the two particles attract each other

**$\sigma$** = distance at which the intermolecular potential between the two particles is zero.  $\sigma$ gives a measurement of how close two nonbonding particles can get and is thus referred to as the van der Waals radius. It is equal to one-half of the internuclear distance between nonbonding particles

**$r$** = distance of separation between both particles (measured from the center of one particle to the center of the other particle)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

epsilon = 0.36
sigma = 2.88



def U_LJ(epsilon, sigma, r):
    return 4 * epsilon * ((sigma / r)**12 - (sigma / r)**6)

r = np.linspace(1*10**-10, 20, 100)
potential = U_LJ(epsilon, sigma, r)

plt.plot(r, potential, 'b')
plt.xlabel('r')
plt.ylabel('U[$\epsilon$, $\sigma$, r]')
plt.title('Lennard Jones Potential')
plt.xlim(0, 20)
plt.ylim(-0.5, 1)
plt.grid(True)
plt.show()


```

### Compare 2 different well depth (epsilon) and equilibrium distance (sigma)

```python
epsilon1 = 0.36  # Well Depth 1
sigma1 = 2.88  # Equilibrium Distance 1


epsilon2 = 0.36  # Well Depth 2
sigma2 = 2.88  # Equilibrium Distance 2



def U_LJ(epsilon, sigma, r):
    return 4 * epsilon * ((sigma / r)**12 - (sigma / r)**6)


plt.plot(r, U_LJ(epsilon1, sigma1, r), 'b', label='Well Depth 1')
plt.plot(r, U_LJ(epsilon2, sigma2, r), 'r', label='Well Depth 2')

plt.xlabel('r')
plt.ylabel('U[$\epsilon$, $\sigma$, r]')
plt.title('Differing Well Depth and Equilibrium Distance')
plt.xlim(0, 20)
plt.ylim(-0.5, 1)
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

##1.2 DOUBLE EXPONENTIAL POTENTIAL

\begin{equation}
U_{ij}^{DE} = \epsilon \left(\frac{\beta e^{\alpha}}{\alpha - \beta} \exp\left(-\alpha \frac{r}{\sigma}\right) - \frac{\alpha e^{\beta}}{\alpha - \beta} \exp\left(-\beta \frac{r}{\sigma}\right)\right)
\end{equation}
<br>



**$U_{ij}^{DE}$** = intermolecular potential between the two atoms or molecules

$\alpha$ = steepness of the repulsive interation

$\beta$ = decay of the attraction

>>$U_{ij}^{DE}$ = $U_{ij}^{LJ}$ when $\alpha$ = 17.47 ; $\beta$ = 4.1

**$\epsilon$** = well depth and a measure of how strongly the two particles attract each other

**$\sigma$** = distance at which the intermolecular potential between the two particles is zero.  $\sigma$ gives a measurement of how close two nonbonding particles can get and is thus referred to as the van der Waals radius. It is equal to one-half of the internuclear distance between nonbonding particles

**$r$** = distance of separation between both particles (measured from the center of one particle to the center of the other particle)

```python
alpha = 17.47
beta = 4.1
sigma = 2.88
epsilon = 0.36

def U_DE(alpha, beta, sigma, epsilon, r):
    return epsilon * ((beta * np.exp(alpha)) / (alpha - beta) * np.exp(-alpha * r / sigma) -
                      (alpha * np.exp(beta)) / (alpha - beta) * np.exp(-beta * r / sigma))



r = np.linspace(0, 20, 10000)
potential =U_DE(alpha, beta, sigma, epsilon, r)

plt.plot(r, potential, 'b')
plt.xlabel('r')
plt.ylabel('U[$\epsilon$, $\sigma$, r]')
plt.title('Double Exponential Potential')
plt.xlim(2.5, 5)
plt.ylim(-0.5, 1)
plt.grid(True)
plt.show()
```
